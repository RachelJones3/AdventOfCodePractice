```python
#I saved the data from Advent of Code Day 1 problem as a .csv file, imported it, then viewed it.
#This problem requires you to take the first and last digit of each row of this data to create a two digit number, then sum the total.
import pandas as pd

data = pd.read_csv('Code Practice Trebuchet.csv')

data
```

```python
#I removed the non-numeric characters with a regular expression.
data['Code'] = data['Code'].str.replace('\D', '', regex = True)
data
```

```python
#I created a function to make a list of all of the first digits then another list of all of the last digits.
def treb_1():
    first_dig = []
    last_dig = []
    for i in data['Code']:
        first_dig.append(i[0])
    for i in data['Code']:
        last_dig.append(i[-1])
#It then combines the two lists to make strings of each two decimal number.
    new_list = [x + y for x, y in zip(first_dig, last_dig)]
#I previewed this to make sure my lists worked as expected.
    print(new_list)
treb_1()
```

```python
#Once I confirmed my lists were working right, I continued this function to sum them together.
def treb_1():
    first_dig = []
    last_dig = []
    for i in data['Code']:
        first_dig.append(i[0])
    for i in data['Code']:
        last_dig.append(i[-1])
    new_list = [x + y for x, y in zip(first_dig, last_dig)]
#Here I change the strings into integers so they could be aggregated.
    for i in range(0, len(new_list)):
        new_list[i] = int(new_list[i])
#Them I printed the sum of this list of integers to get the answer.
    print(sum(new_list))
treb_1()
```
